```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv('summary.csv')

#replace Embedded in data with Co-Clustering On Embedded Data
df['Type'] = df['Type'].replace('Embedded', 'Co-Clustering On Embedded Data')
#replace simple_clustering with Traditional Clustering on Embedded Data
df['Type'] = df['Type'].replace('simple_clustering', 'Traditional Clustering on Embedded Data')
# Aggregate mean MSE for all methods
mean_mse = df.groupby(['Technique', 'Type'])['MSE'].mean().reset_index()

# Plot the mean MSE for all aggregated methods
plt.figure(figsize=(14, 7))
sns.barplot(x='Type', y='MSE', hue='Technique', data=mean_mse, palette='viridis')
plt.xlabel('Type')
plt.ylabel('Mean MSE')
plt.title('Mean MSE for All Aggregated Methods')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('mean_mse_all_methods.png')
plt.show()

# Sort the dataframe by MSE and get the top 10 approaches
top_10 = df.nsmallest(10, 'MSE')

# Plot the MSE of the top 10 approaches
plt.figure(figsize=(14, 7))
sns.barplot(x='Type', y='MSE', hue='Technique', data=top_10, palette='viridis', dodge=False)
plt.xlabel('Type')
plt.ylabel('MSE')
plt.title('MSE of the Top 10 experiments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Technique')
plt.tight_layout()
plt.savefig('top_10_mse_approaches.png')
plt.show()
```
